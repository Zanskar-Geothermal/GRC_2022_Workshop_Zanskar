```python
#How does PCA work? 

%%html
<iframe src="https://drive.google.com/file/d/1ReGIafspdGDxsAqRU0eNB6kbq3FH33ac/preview" width="640" height="480" allow="autoplay"></iframe>
```

PCA is a common tool for simplifying mutlidimensional data (e.g., 10D) into smaller dimensions (e.g., 2D) to support clustering. Kmeans clustering randomly initializes centroid locations and iteratively refines these locations until it finds an optimal centroid for each cluster group

```python
from sklearn.decomposition import PCA
n = len(X.columns)
pcamodel = PCA(n_components=n)
pca = pcamodel.fit_transform(X)

#Index the scores from the PCA reconstruction
score = pca[:,0:2]

#Create a new dataframe with all features and PC scores

df_pcaTsite = pd.DataFrame(pca,columns=[ "PC"+str(X) for X in range(1,pcamodel.n_components_+1)])
df_pca = pd.concat([df,df_pcaTsite], axis=1, sort=False)
```

```python
#ADD ELLIPSES
from sklearn.cluster import KMeans

from scipy.spatial.distance import cdist

nCluster = 3
"""The radius of each circle is the distance between the centroid and points in
each cluster, either the min, mean, or max dist of the collection of points"""

def plot_kmeans(kmeans, X, Xneg, Xpos, n_clusters=nCluster, rseed=0, ax=None):
    labels = kmeans.fit_predict(X)


    labels_Xneg = labels[0:62]
    labels_Xpos = labels[62:145]

    # plot the input data as scatter points
    ax = ax or plt.gca()

    label1 = 'Positive T-Site'
    label2 = 'Negative T-Site'
    ax.scatter(Xneg[:, 0], Xneg[:, 1], c=labels_Xneg, s=85, edgecolors='k', marker='o', cmap='rainbow', label = label1, zorder=2)
    ax.scatter(Xpos[:, 0], Xpos[:, 1], c=labels_Xpos, s=285,marker=(5, 1), edgecolors='k', cmap='rainbow',label = label2, zorder=2)


    plt.grid()

    # plot ellipse representations of the KMeans clusters
    centers = kmeans.cluster_centers_ 
    radii = [cdist(X[labels == i], [center]).mean()
             for i, center in enumerate(centers)]
    for c, r in zip(centers, radii):
          ax.add_patch(plt.Circle(c, r, fc='white',ec ='cyan', lw=3, alpha = 0.5, zorder=1))

    plt.scatter(centers[:,0] , centers[:,1] , s = 380, marker='o', edgecolors='k',color = 'white')
    for i in range(centers.shape[0]):
      plt.text(centers[i,0], centers[i,1], str(i+1),fontsize=12,fontweight='bold',horizontalalignment='center', verticalalignment='center')

    ncols = 1
    plt.legend(scatterpoints=1, ncol=ncols, loc='center left',bbox_to_anchor=(1, 0.5),fontsize=20)
```

```python
#ADD VECTORS

"""The loading vectors of each feature illustrate their 
  influence on the variance of our data"""

def vector_plot(score,coeff,labels=None):
    angle = 20
    var_ratio1 = float(str(round(pcamodel.explained_variance_ratio_[0]*100, 2)))
    var_ratio2 = float(str(round(pcamodel.explained_variance_ratio_[1]*100, 2)))
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0]*10, coeff[i,1]*10,color = 'k',alpha = 0.5, lw = 2 ) #MULTIPLY BY 10
        if labels is None:
            plt.text(coeff[i,0]* 1.05, coeff[i,1] * 1.05, "Var"+str(i+1), color = 'k', ha = 'center', va = 'center')

        else:
            plt.text(coeff[i,0]* 10.00, coeff[i,1] * 10.00, labels[i],color = 'k',horizontalalignment='center', verticalalignment='center_baseline', rotation=angle, rotation_mode='anchor', fontsize=12,fontweight='bold')
 
    plt.xlabel('PC1 ' + str(var_ratio1) + '%',fontsize=20)
    plt.ylabel('PC2 ' + str(var_ratio2) + '%',fontsize=20)
```

```python
#Organize principal components by train code 
df_pca = df_pca.sort_values(by='TrainCode',ascending = True)

df_pca_neg = df_pca[df_pca['TrainCode']==0] 
df_pca_pos = df_pca[df_pca['TrainCode']==1] 

df_pca_pos = df_pca_pos[['PC1','PC2']]
df_pca_pos = df_pca_pos.to_numpy()
df_pca_neg = df_pca_neg[['PC1','PC2']]
df_pca_neg = df_pca_neg.to_numpy()

#Isolate only the first PC values for plot_kmeans function 
df_pca_first_two = df_pca[['PC1','PC2']]
df_pca_first_two = df_pca_first_two.to_numpy()
```

Siler and Pepin 2018 explanation of a PC biplot

```python
n_clusters =  [2,3,4,5]

if IN_COLAB:
    tb3 =  widgets.TabBar([str(n_clusters[i]) +' Clusters ' for i in range(np.shape(n_clusters)[0])])
else:
  fig, myax = plt.subplots(figsize=[10,15],ncols = 2, nrows = np.shape(n_clusters)[0])

for i in range(np.shape(n_clusters)[0]):  
  
  with tb3.output_to(i, select=(i < 2)):

    #Call the kmeans plotting function 
    kmeans = KMeans(n_clusters=i+2, random_state=42)
    plot_kmeans(kmeans,df_pca_first_two,df_pca_neg,df_pca_pos) 

    #Call the loading vector plotting function 
    vector_plot(pca[:,0:2],np.transpose(pcamodel.components_[0:2, :]),list(X.columns))
    plt.xlim([-4, 6])
    plt.ylim([-6, 10]) 
    plt.tight_layout()
```

We can visually assume that ~3 works best for the data. We can also look at a metric like within-group sum of squares 

which is this calculation
 <font size="5">   
 $\ ss = \Sigma_{j=1}^{K} \Sigma_{i=1}^{l} (\bar{X}(i) - \bar{X}(j))^2$ 
    </font>

Representing the differences among group means (i.e. distortion)

```python
fig= plt.figure(figsize=(7,5)) 

#Append distortions to list
distortions = []
for i in range(1, 9): #test cluster values from 1-9
    km = KMeans(n_clusters=i, init='k-means++') 
    #Fit K-means to the first PCs which explain the most variance
    km.fit(df_pca_first_two)
    distortions.append(km.inertia_)
K = range(1, 9)

plt.plot(K,distortions,marker='o')
plt.grid()
plt.xlabel('Number of clusters')
plt.ylabel('Within-Group Sum of Squares')
plt.title('Clusters (K) vs Distortion',fontweight='bold')
plt.tight_layout ()    
```
